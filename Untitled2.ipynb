```python
class myStack:
    # constructor
    def __init__(self,cap):
        self.capacity=cap
        self.arr=[0]* self.capacity
        self.top=-1

    #push operation
    def push(self,x):
        if self.top == self.capacity - 1:
          print("stack overflow")
          return
        self.top +=1
        self.arr[self.top]=x

    #pop operation
    def pop(self):
        if self.top==-1:
            print("stack underflow")
            return -1
        val=self.arr[self.top]
        self.top -=1
        return val

    #peek operation
    def peek(self):
        if self.top == -1:
            print("stack is empty")
            return -1
        return self.arr[self.top]
        
    #check if stack is empty
    def isEmpty(self):
      return self.top ==-1

    #check if stack is full
    def isFull(self):
      return self.top == self.capacity - 1

if __name__ == "__main__":

        #get stack size from user
        size=int(input("enter the size of the stack"))

        #create stack
        st= myStack(size)

        #get number of elements
        n=int(input("enter the number of elements to push:"))

        #push elements
        for i in range(n):
            x=int(input("enter element {i+1}:"))
            st.push(x)

        #pop one element
        print("popped",st.pop())

        #display top element
        print("top element", st.peek())

        #check if stack is empty
        print("is stack empty:", "yes" if st.isEmpty() else "no")

        #check if stack is full
        print("is stack full:", "yes" if st.isFull() else "no")

    

    

    
        
```
